```python
import os
import shutil

# --- Configuration ---
source_dir = "./dataset/PMC dataset/2D/validation/1.5T/T1/"   # your input folder
target_dir = "./dataset/PMC dataset/2D/validation/1.5T/T1_modified/"     # your output folder

# create target folder if it doesn’t exist
os.makedirs(target_dir, exist_ok=True)

# loop through each file in source folder
for filename in os.listdir(source_dir):
    if not filename.lower().endswith((".jpg", ".jpeg", ".png", ".tif", ".tiff")):
        continue  # skip non-image files

    if "1o5" in filename:   # check if substring exists
        new_filename = filename.replace("1o5", "3")  # rename pattern
        src_path = os.path.join(source_dir, filename)
        dst_path = os.path.join(target_dir, new_filename)

        # copy the file to new folder with renamed filename
        shutil.copy2(src_path, dst_path)
        print(f"✅ Renamed and copied: {filename} → {new_filename}")
    else:
        print(f"Skipped (no 1o5 found): {filename}")

print("\n🎉 Done! All files renamed and copied successfully.")
```

```python
import os
from PIL import Image
import math

# Path to your combined dataset
dataset_root = "/srv/Data16tb/siva/ssiva/MR_Translation/modified_dataset/destination_3T"

# Folders to process
folders = ["train", "val", "test"]

for folder in folders:
    folder_path = os.path.join(dataset_root, folder)
    if not os.path.exists(folder_path):
        print(f"Folder {folder_path} does not exist, skipping...")
        continue

    print(f"Processing folder: {folder_path}")
    for filename in os.listdir(folder_path):
        if not filename.lower().endswith((".png", ".jpg", ".jpeg")):
            continue

        img_path = os.path.join(folder_path, filename)
        img = Image.open(img_path)
        w, h = img.size

        # Calculate new height (next multiple of 4)
        new_h = math.ceil(h / 4) * 4
        pad_top = (new_h - h) // 2
        pad_bottom = new_h - h - pad_top

        if new_h != h:
            # Apply padding (top/bottom), width remains unchanged
            padded_img = Image.new(img.mode, (w, new_h))
            padded_img.paste(img, (0, pad_top))
            padded_img.save(img_path)
            print(f"Padded {filename}: {h} → {new_h}")
        else:
            print(f"{filename} already divisible by 4, skipping...")

print("✅ All images processed successfully!")
```

```python
import os
from PIL import Image

# Root dataset folder
dataset_root = "/srv/Data16tb/siva/ssiva/MR_Translation/modified_dataset/aligned_dataset"

# Folders to process
folders = ["train", "val", "test"]

# Target dimensions
TARGET_WIDTH = 512
TARGET_HEIGHT = 152  # target height

for folder in folders:
    folder_path = os.path.join(dataset_root, folder)
    if not os.path.exists(folder_path):
        print(f"[WARN] Folder {folder_path} does not exist, skipping...")
        continue

    print(f"\nProcessing folder: {folder_path}")

    for fname in os.listdir(folder_path):
        if not fname.lower().endswith((".png", ".jpg", ".jpeg")):
            continue

        img_path = os.path.join(folder_path, fname)
        try:
            img = Image.open(img_path)
            mode = img.mode  # preserve original mode (likely 'L' for grayscale)
            w, h = img.size

            # Only process if height is different from target
            if h != TARGET_HEIGHT:
                if h < TARGET_HEIGHT:
                    # Pad top/bottom evenly
                    pad_top = (TARGET_HEIGHT - h) // 2
                    pad_bottom = TARGET_HEIGHT - h - pad_top
                    new_img = Image.new(mode, (w, TARGET_HEIGHT))
                    new_img.paste(img, (0, pad_top))
                    img = new_img
                    print(f"Padded {fname}: height {h} -> {TARGET_HEIGHT}")
                elif h > TARGET_HEIGHT:
                    # Crop vertically
                    top = (h - TARGET_HEIGHT) // 2
                    img = img.crop((0, top, w, top + TARGET_HEIGHT))
                    print(f"Cropped {fname}: height {h} -> {TARGET_HEIGHT}")

                # Width should remain untouched
                # Save image in **original mode** to preserve grayscale
                img.save(img_path)

        except Exception as e:
            print(f"[ERROR] Failed processing {fname}: {e}")

print("\n✅ All folders processed. All images should now have uniform height and original properties preserved.")
```
